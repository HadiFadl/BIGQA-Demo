```python

import json
from kafka import KafkaProducer
import time
import threading
```

```python
def producer_instance(kafka_brokers):
        
    try:
        producer = KafkaProducer(bootstrap_servers=kafka_brokers,value_serializer=lambda x: json.dumps(x).encode('utf-8'))
    except Exception as msg :
            print('The following exception was raised when connecting to Kafka: {}'.format(msg))

    return producer

def yield_data(filename):

    file_content = [json.loads(line) for line in open(filename, 'r' , encoding='utf-8')]
    for response_line in file_content:
        time.sleep(0.5)
        if response_line:
            yield response_line

def kafka_produce(yield_data, filename, producer_instance,kafka_topic) -> None:
        
    kafka_topic = kafka_topic
    producer = producer_instance
    yielder = yield_data(filename)

    for one_mesurment in yielder:
        producer.send(kafka_topic, one_mesurment)

def sensor_sim1():
    while True:
        kafka_produce(yield_data, filename ='sensornull1.json', producer_instance = producer, kafka_topic ='sensor')
        print("starting prod")

def sensor_sim2():
    while True:
        kafka_produce(yield_data, filename ='sensornull2.json', producer_instance = producer, kafka_topic ='sensor')
        print("starting prod")

def sensor_sim3():
    while True:
        kafka_produce(yield_data, filename ='sensornull3.json', producer_instance = producer, kafka_topic ='sensor')
        print("starting prod")

```

```python
kafka_brokers = 'localhost:9092'
producer = producer_instance(kafka_brokers)
```

```python
def main():
    
    # Create threads for each function
    t1 = threading.Thread(target=sensor_sim1)
    t2 = threading.Thread(target=sensor_sim2)
    t3 = threading.Thread(target=sensor_sim3)

    # Start the threads
    t1.start()
    t2.start()
    t3.start()

    # Wait for all threads to finish
    t1.join()
    t2.join()
    t3.join()

if __name__ == '__main__':
    main()
```
